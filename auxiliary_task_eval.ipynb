```python
def S(x, y):
    if y < x:
        return 1
    elif x == y:
        return 0.5
    else:
        return 0
    
def probability_of_improvement(x_scores, y_scores):
    N = len(x_scores)  # Jumlah percobaan untuk X
    K = len(y_scores)  # Jumlah percobaan untuk Y
    
    total_sum = 0
    for i in range(N):
        for j in range(K):
            total_sum += S(x_scores.iloc[i], y_scores.iloc[j])
    
    # Probability of improvement
    return total_sum / (N * K)

P_improvement = probability_of_improvement(no_aux_game_df["block_placed"], full_aux_game_df["block_placed"])
P_improvement
```
